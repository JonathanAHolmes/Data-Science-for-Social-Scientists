# In-Class Exercise — Part 1: Dummy Variables

A business owner models seasonal sales as:

$$Y = \beta_0 + \beta_1 X_1 + \beta_2 X_2 + \beta_3 X_3 + u_i$$

where $X_1$ = fall, $X_2$ = winter, $X_3$ = spring are dummy variables (summer is the **omitted/reference** category). Estimated coefficients: $\hat{\beta}_0 = 100$, $\hat{\beta}_1 = 10$, $\hat{\beta}_2 = -10$, $\hat{\beta}_3 = 50$.

## Question 1

**What were average sales in fall, winter, spring, and summer?**

---

**Answer.**

For each season, set the corresponding dummy to 1 and all others to 0:

| Season | Calculation | Average Sales |
|--------|------------|---------------|
| **Summer** | $\hat{\beta}_0 = 100$ (all dummies = 0) | **100** |
| **Fall** | $\hat{\beta}_0 + \hat{\beta}_1 = 100 + 10$ | **110** |
| **Winter** | $\hat{\beta}_0 + \hat{\beta}_2 = 100 - 10$ | **90** |
| **Spring** | $\hat{\beta}_0 + \hat{\beta}_3 = 100 + 50$ | **150** |

The intercept $\hat{\beta}_0 = 100$ is the predicted (average) value for the omitted category, **summer**. Each slope coefficient measures the *difference* in average sales relative to summer.

## Question 2

**Suppose I re-estimated with fall as the omitted category:**

$$Y = \alpha_0 + \alpha_2 X_2 + \alpha_3 X_3 + \alpha_4 X_4 + u_i$$

**Is it possible to predict the new estimates? What are they?**

---

**Answer.**

Yes — the **fitted values** (predicted averages for each season) are fixed by the data and do not change when we choose a different reference category. Only the *interpretation* of the coefficients changes.

With fall as the omitted category, the intercept becomes the average sales in fall, and each slope measures the difference relative to fall:

| Coefficient | Meaning | Calculation | Value |
|------------|---------|-------------|-------|
| $\hat{\alpha}_0$ | Average sales in fall | same as fall's average | **110** |
| $\hat{\alpha}_2$ | Winter $-$ Fall | $90 - 110$ | **$-20$** |
| $\hat{\alpha}_3$ | Spring $-$ Fall | $150 - 110$ | **$+40$** |
| $\hat{\alpha}_4$ | Summer $-$ Fall | $100 - 110$ | **$-10$** |

> The **predicted value for any observation is identical** across both specifications — the choice of reference category affects only the *labels* on the coefficients, not the fit of the model.

# In-Class Exercise — Part 2: Classification and ROC Curves

Three classifiers (logistic regression, LDA, and a linear probability model) are compared using ROC curves on the `Default` dataset.

## Question 3

**If I want to target a 0.4 false-positive rate, which model gives the best predictions?**

---

**Answer.**

On an ROC curve, the **false-positive rate (FPR)** is on the $x$-axis and the **true-positive rate (TPR)** is on the $y$-axis. For a target FPR of 0.4, we read up from $x = 0.4$ and compare the TPR achieved by each model at that point.

The **best model** at FPR = 0.4 is whichever curve sits **highest** at $x = 0.4$ — it catches the most true positives for the same false-alarm rate. In the class demonstration, logistic regression and LDA tend to overlap closely and both outperform the linear probability model, so the winner is typically **logistic regression or LDA** (consult the actual plot to confirm which is higher at exactly 0.4).

## Question 4

**Is one of these three models definitively "the best"? Why or why not?**

---

**Answer.**

No. The ROC curve shows that **no model dominates at every operating point**. The "best" model depends on the **loss function** — specifically, the relative cost of false positives versus false negatives:

* If false positives are very costly (e.g., a bank wrongly denying good customers), we want to operate at a low FPR, and the model with the highest TPR at that low FPR is best.
* If missing true positives is more costly (e.g., failing to flag a defaulter), we want high TPR even at the expense of more false positives.

Moreover, this comparison is on the **training/in-sample data**. Without a held-out test set, we cannot be sure which model will generalise best — the in-sample ROC curves can be optimistic.

## Question 5

**Why is the linear probability model doing an okay job here, even though we know the binned scatterplot fit is terrible?**

---

**Answer.**

The ROC curve evaluates a model by its **ranking** of observations by predicted probability, not by the absolute accuracy of those probabilities. Even if the linear probability model assigns predicted values outside $[0,1]$ and its fitted line looks poor in the scatterplot, it may still **rank observations in roughly the right order** — placing likely defaulters above unlikely ones.

ROC-AUC (area under the curve) is a **rank-based** metric: as long as the model orders observations correctly (higher score → more likely to be positive), it will have a high AUC regardless of whether the raw scores are well-calibrated probabilities. The linear model's poor calibration hurts tasks like computing precise default probabilities, but not necessarily the classification ranking task.

## Question 6

**Would having a second "test" dataset be helpful in any way? If so, how?**

---

**Answer.**

Yes, a held-out **test set** is essential for several reasons:

1. **Honest model comparison:** ROC curves constructed on the training data are optimistic — models have seen these observations and can fit them well even by memorising noise. A test set reveals how well each model actually generalises to new data.
2. **Model selection:** We can compare test AUC (or test TPR at a target FPR) across the three classifiers and choose the one that truly performs best, rather than the one that fits the training data best.
3. **Threshold selection:** We can validate that the chosen decision threshold (e.g., cutting at predicted probability 0.5, or at FPR 0.4) performs as expected on unseen observations, rather than being an artifact of the training sample.

In practice, when a separate test set is unavailable, cross-validation can serve the same purpose.
