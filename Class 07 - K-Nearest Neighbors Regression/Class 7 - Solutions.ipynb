## Question 1

**In the "1 predictor vs 2 predictors" demo we compared KNN with TV only vs TV + Radio (scaled). Why do we scale when using more than one predictor? What would go wrong if we did not scale?**

---

**Answer.**

KNN defines "nearest" using a **distance metric** (typically Euclidean distance). For two observations $i$ and $j$ with predictors $X_1$ and $X_2$:

$$d(i, j) = \sqrt{(X_{1i} - X_{1j})^2 + (X_{2i} - X_{2j})^2}$$

If the predictors are measured on very different scales — say TV budgets range from 0 to 300 (\$thousands) while Radio ranges from 0 to 50 — then differences in TV will numerically dominate the distance, even if Radio carries more predictive information per unit. The algorithm would effectively **ignore the small-scale predictor**.

Standardising each predictor to zero mean and unit variance (z-scores) puts all variables on an equal footing so that each predictor contributes to distance in proportion to its variability, not its arbitrary unit of measurement.

## Question 2

**In the "Unscaled vs scaled" demo, KNN with scaled TV + Radio had lower test MSE than with unscaled. In one sentence, what is the main reason scaling helped? Do you think scaling will always help the fit?**

---

**Answer.**

**Why scaling helped:** Scaling put TV and Radio on equal numerical footing, so the distance metric used genuine information from both predictors rather than being dominated by TV's larger raw values.

**Will scaling always help?** Not necessarily. If one predictor genuinely has far more predictive power than another, letting it dominate distance calculations might actually be appropriate. Scaling always makes the algorithm treat predictors symmetrically — that is beneficial when predictors are equally relevant but could in principle mask signal if one predictor is truly more informative *per unit of its natural scale*. In practice, however, scaling is almost always recommended before applying any distance-based method.

## Question 3

**When would you prefer linear regression over KNN regression for predicting a quantitative response? When might you prefer KNN?**

---

**Answer.**

**Prefer linear regression when:**

* The true relationship between $Y$ and $X$ is approximately **linear** (or can be made linear with simple transformations). Linear regression will have lower variance than KNN and will not waste observations fitting local curvature that does not exist.
* **Interpretability** is important — linear regression gives clear coefficient estimates with standard errors, confidence intervals, and $p$-values.
* The dataset has **many predictors relative to observations** (high-dimensional settings). KNN suffers severely from the curse of dimensionality: the neighbourhood around any point becomes sparse in high dimensions, so the $K$ nearest neighbours are no longer truly "near."
* **Extrapolation** is needed — KNN cannot predict outside the range of the training data.

**Prefer KNN when:**

* The true $f(X)$ is **nonlinear** in ways that are hard to specify in advance (e.g., complex interactions, thresholds).
* You have a **moderate number of predictors** and enough observations to populate neighbourhoods adequately.
* **Flexibility** matters more than interpretability, and you are willing to tune $K$ via cross-validation.
