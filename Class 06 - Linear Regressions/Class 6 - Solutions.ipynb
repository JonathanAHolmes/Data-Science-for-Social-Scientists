# In-Class Exercise #1 — Test Your Econometrics Knowledge!

The regression output below comes from regressing **Sales** on **TV** advertising spend using the `Advertising.csv` dataset ($n = 200$ markets).

|   | coef | std err | t | P>\|t\| | [0.025 | 0.975] |  
| --- | :-: | :-: | :-: | :-: | :-: | :-: |
| Intercept | 7.0326 | 0.458 | 15.360 | 0.000 | 6.130 | 7.935 |
| TV | 0.0475 | 0.003 | 17.668 | 0.000 | 0.042 | 0.053 |

## Question 1

**What are the independent and dependent variables? How many observations are there?**

---

**Answer.**

* **Dependent variable ($Y$):** Sales (thousands of units).
* **Independent variable ($X$):** TV advertising budget (thousands of dollars).
* **Number of observations:** $n = 200$ (200 markets in the `Advertising.csv` dataset).

## Question 2

**Identify each of the following from the table.**

---

**Answer.**

| Quantity | Value |
|---|---|
| $\hat{\beta}_0$ (intercept) | **7.0326** |
| $\hat{\beta}_1$ (slope on TV) | **0.0475** |
| Standard error of $\hat{\beta}_1$ | **0.003** |
| 95% confidence interval of $\hat{\beta}_1$ | **[0.042, 0.053]** |
| 95% confidence interval of $\hat{\beta}_0$ | **[6.130, 7.935]** |

Interpretation of $\hat{\beta}_1$: for every additional \$1,000 spent on TV advertising, expected sales increase by **47.5 units** on average.

## Question 3

**Is more TV spending associated with higher sales? What do you look at in the table to figure this out?**

---

**Answer.**

Yes. There are two things to check:

1. **Sign of the coefficient:** $\hat{\beta}_1 = 0.0475 > 0$, so the relationship is positive — more TV spending is associated with higher sales.
2. **Statistical significance:** the $p$-value for TV is $0.000 < 0.05$, and the 95% confidence interval $[0.042, 0.053]$ does not include zero. We can therefore reject $H_0: \beta_1 = 0$ and conclude the association is statistically significant.

> **Caution:** this is an observational association. We cannot conclude that TV spending *causes* higher sales without additional assumptions (e.g., an experiment or instruments).

## Question 4

**What does the column $P>|t|$ mean?**

---

**Answer.**

The column $P>|t|$ is the **$p$-value** for a two-sided $t$-test of the null hypothesis $H_0: \beta_j = 0$.

Formally, it is the probability of observing a $t$-statistic at least as extreme as the one computed from the data, *assuming the null hypothesis is true*:

$$p = P\!\left(|T| \geq |\hat{t}| \mid H_0\right), \qquad \hat{t} = \frac{\hat{\beta}_j}{\widehat{\text{SE}}(\hat{\beta}_j)}$$

A small $p$-value (conventionally $< 0.05$) means the data are unlikely under $H_0$, so we **reject** the null and conclude $\beta_j \neq 0$. A large $p$-value means we **fail to reject** — the data are consistent with the predictor having no effect.
