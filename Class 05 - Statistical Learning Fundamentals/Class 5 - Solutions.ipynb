# In-Class Exercise #1

Recall the two different fire alarm algorithms.

Algorithm A: when there is no fire the alarm rings 5% of the time; when there is a fire it rings 99% of the time.

Algorithm B: when there is no fire the alarm rings 2% of the time; when there is a fire it rings 90% of the time.

## Q1

**Which of these two algorithms has a lower mean squared error? (Reminder: fires are rare.)**

---

**Answer.**

Because the outcome is binary (fire = 1, no fire = 0), the mean squared error equals the overall misclassification rate. Define the base rate of fires as $p$ (small, since fires are rare — say $p = 0.01$). The MSE for each algorithm is:

$$\text{MSE}_A = p \times (1 - 0.99) + (1-p) \times 0.05 = 0.01 \times 0.01 + 0.99 \times 0.05 \approx 0.0496$$

$$\text{MSE}_B = p \times (1 - 0.90) + (1-p) \times 0.02 = 0.01 \times 0.10 + 0.99 \times 0.02 \approx 0.0208$$

**Algorithm B has the lower MSE.** Even though it misses more fires (10% miss rate vs 1%), fires are so rare that the false-positive rate dominates the overall error. Algorithm A triggers unnecessary alarms 5% of the time it should be silent — and since "no fire" is the common case, this inflates its total error substantially.

## Q2

**When should we try to minimize mean squared error, and when is it appropriate to have a different goal?**

---

**Answer.**

MSE (or misclassification rate) treats all errors as equally costly. This is appropriate when false positives and false negatives have roughly equal consequences.

A different objective is warranted when the **costs of errors are asymmetric**:

* **Missing a fire is far more costly than a false alarm.** An occupant who dies because the alarm failed to ring is a far worse outcome than an unnecessary evacuation. In this case we should weight false negatives heavily — maximising the true-positive rate (sensitivity) matters more than minimising overall errors. A loss function that heavily penalises missed fires is more appropriate than raw MSE.

* **False alarms are very costly.** In some settings (e.g., a hospital surgery suite where every alarm triggers an expensive response), frequent false positives cause real harm. Then we might instead target a very low false-positive rate and accept some missed events.

More generally, MSE is just one loss function. Any time the real-world consequences of different error types differ — spam detection, medical diagnosis, fraud flagging, loan approvals — the loss function should reflect those costs rather than defaulting to MSE.

# In-Class Exercise #2

After fitting the true function $f(x) = \cos(1.5\pi x)$ to the 30-observation sample, the MSE is not zero.

## Q3

**Why is the MSE of the true function not zero?**

---

**Answer.**

The data were generated as $y_i = f(x_i) + \varepsilon_i$ where $\varepsilon_i \sim N(0, 0.1^2)$ is random noise. Even when we use the exact true function $f(x) = \cos(1.5\pi x)$, the predictions $\hat{y}_i = f(x_i)$ differ from the observed $y_i$ by $\varepsilon_i$. The MSE of the true function therefore estimates:

$$\text{MSE}_{\text{true}} \approx \frac{1}{n}\sum_{i=1}^n \varepsilon_i^2 \xrightarrow{n\to\infty} E[\varepsilon^2] = \text{Var}(\varepsilon)$$

This is the **irreducible error** — the component of prediction error that cannot be eliminated no matter how good our model is. It exists because the data-generating process itself is stochastic. Even a perfect model cannot predict noise.

## Q4

**How do you expect the MSE of the true function to change as we increase the sample size $N$?**

---

**Answer.**

The MSE of the true function **converges to a fixed value** — it does not systematically fall toward zero as $N$ grows.

By the law of large numbers:

$$\frac{1}{N}\sum_{i=1}^N \varepsilon_i^2 \xrightarrow{p} E[\varepsilon^2] = \text{Var}(\varepsilon) = 0.01$$

With a small sample ($N = 30$), the sample MSE is a noisy estimate of this quantity and may be somewhat above or below 0.01 due to sampling variation. As $N$ increases, that sampling variation shrinks — the MSE estimate becomes more precise — but it converges to $\text{Var}(\varepsilon)$, not to zero.

This is the defining property of irreducible error: **more data cannot help**. We can reduce the *reducible* part of MSE by improving our model, but we can never reduce the irreducible part.

## Q5

**What degree of polynomial would you choose to give the best predictions $\hat{y}$ in this example?**

---

**Answer.**

The best degree is the one that **minimises test (out-of-sample) MSE**, not training MSE.

As we increase the polynomial degree:

* **Training MSE** falls monotonically — a higher-degree polynomial can always fit the training points more closely.
* **Test MSE** follows a U-shaped curve: it first falls as the model captures real structure in the data, then rises once the model starts fitting noise (overfitting).

The optimal degree sits at the **bottom of the test-MSE curve**. In the class demonstration the true function is $\cos(1.5\pi x)$, which is well approximated by a low-degree polynomial (roughly degree 3–5). A degree-1 fit is **under-fit** — high bias; degrees beyond roughly 5–6 are **over-fit** — high variance. The train/test MSE plot reveals the crossover point directly.

**Practical rule:** use a held-out test set (or cross-validation, introduced in Class 9) to plot test MSE against polynomial degree, then pick the degree at the minimum.

> This is the fundamental **bias–variance tradeoff**: flexibility reduces bias but increases variance. The optimal model balances the two.
