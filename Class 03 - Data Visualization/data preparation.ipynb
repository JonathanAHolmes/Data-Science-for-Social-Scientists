```python
# import our packages and use aliases
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
pathFolder="~/Dropbox/Adaptation_Completeness/Data/VDSA/"
class3Folder="~/Dropbox/Data Science for Social Scientists/Classes/Class 3 - Data Visualization/"
```

```python
df=pd.read_stata(pathFolder+"vdsa_data_by_district_crop_year.dta")
df=df.loc[df.year.between(1995,2015)]
df.set_index(['statename','distname','statecode','distcode','year','crop'],inplace=True)
df=df[['yield','price','area']]
crops=['rice', 'wheat', 'sorghum', 'maize', 'pearlmillet']
df=df.query("crop==@crops")
df=df.unstack(level=-1)
df.columns = df.columns.map('_'.join)
df.reset_index(inplace=True)
for c in ['price_maize','price_pearlmillet','price_rice','price_sorghum','price_wheat']:
    df[c]= df[c].astype(str)
    df[c]= df[c].str.replace("nan","")
    df.loc[df[c]!="",c]= "R "+df[c]
for c in ['yield_maize','yield_pearlmillet','yield_rice','yield_sorghum','yield_wheat','area_maize','area_pearlmillet','area_rice','area_sorghum','area_wheat']:
    df.loc[df[c].isna(),c]=-1
df.to_stata(class3Folder+"vdsa.dta",write_index=False)
df.head()
```
