# In-Class Exercise #1

**Q1. Give an example of some data that exists. Ideally, this is data you could access, but that is not necessary. Identify a dependent variable you care about from that data.**

---

**Answer (example).**

**Dataset:** the Canadian *Labour Force Survey* (LFS), a monthly Statistics Canada survey of about 100,000 Canadians 15 and older. For each respondent, the LFS records demographics (age, sex, province, education, family status), labour-market status (employed, unemployed, not in the labour force), occupation, industry, and — for paid employees — hourly and weekly wages.

**Dependent variable I care about:** the respondent's **hourly wage** (in dollars per hour).

Wages are interesting because they vary widely across people in ways that are partly explained by observable characteristics (education, experience, occupation, region) and partly by things that are hard to measure (ability, effort, networks, bargaining position). That makes wages a natural target for both descriptive, predictive, and causal questions — which is what Q2 is about.

> *Other equally good examples:* second-hand car prices on AutoTrader (dependent variable: sale price); restaurant inspection records (dependent variable: pass/fail); your own academic grades over time (dependent variable: course grade); the Ottawa public-library catalogue (dependent variable: number of times a book is checked out).

**Q2. Using the data you discussed in Q1, give one example of each of the three types of questions:**
* **Descriptive**
* **Predictive**
* **Causal**

---

**Answer (example, for the LFS hourly-wage example).**

* **Descriptive:** *What is the median hourly wage of full-time employed Canadians aged 25–54 in 2025, and how does it differ across provinces?* This question simply summarizes patterns in the data. There is no model and no claim about *why* wages differ — just a statement of what the data show.

* **Predictive:** *Given a new respondent's age, education, occupation, industry, and province, what hourly wage would we expect them to earn?* This question uses the data to build a model that maps observable characteristics into a forecast for an outcome we have not yet seen. We care about the accuracy of the prediction, not whether the inputs *cause* the wage.

* **Causal:** *If we increased the minimum wage in Ontario by \$2/hour, how would the hourly wages of low-wage workers change?* This question asks about the effect of an *intervention* — changing one input while everything else stays the same. Answering it convincingly requires more than correlations: we need a research design (e.g., a comparison of provinces that did and did not raise their minimum wage) that rules out the many other reasons wages might also have changed.

The same dataset can support all three types of questions, but each requires a different way of thinking about what the data can and cannot tell you.

**Question: What methods would you use to answer each type of question?**

---

**Answer.**

Roughly:

| Type of question | Typical methods |
|------------------|-----------------|
| **Descriptive** | Summary statistics (means, medians, percentiles), cross-tabs, plots (histograms, scatter plots, time series). The point is to *describe* what is in the data, so the methods are simple and the focus is on clear visualization. |
| **Predictive** | Statistical-learning models that map inputs $X$ to a forecast $\hat Y$: linear regression, logistic regression, KNN, decision trees, random forests, boosting, neural networks. We evaluate them with held-out test error or cross-validation, and pick the model that generalizes best. (This is most of what we will do in this course.) |
| **Causal** | An identification strategy that goes beyond correlation: randomized experiments (the gold standard), natural experiments, instrumental variables, difference-in-differences, regression discontinuity, panel methods with fixed effects. The hard part is not the statistics — it is convincing yourself (and others) that you have isolated the effect of the intervention from everything else that might co-vary with it. |

An important caveat for this course: a model that *predicts* well is not automatically a model that *explains causally*. Most of the methods we will learn (KNN, trees, neural networks) are excellent for prediction but cannot, on their own, identify causal effects. Knowing which type of question you are asking determines which set of tools is appropriate.

This is a setup check rather than a conceptual question. Any of the following is a valid choice:
